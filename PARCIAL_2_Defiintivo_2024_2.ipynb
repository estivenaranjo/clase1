x(t) = |A \cos(2\pi F_0 t)|^2, \quad t \in \left[-\frac{1}{2F_0}, \frac{1}{2F_0}\right]
$$, con $$ A, F_0 \in \mathbb{R}^+ $$

Realice las simulaciones respectivas para graficar el espectro de Fourier del ejercicio 1 (magnitud y fase como diagrama de Bode en decibelios), y presente el error relativo y la senal reconstruida para $$ N = \{1, 2, \dots, 50\} $$





## Solución:

Se tiene que:

$$x(t)=|A\cos(2\pi Fo t)|^2 = A^2\cos^2(2\pi Fo t) = A^2\left(\frac{1}{2}+\frac{\cos(2*2\pi Fo t)}{2}\right) = \frac{A^2}{2}+\frac{A^2*\cos(4\pi Fo t)}{2}$$

Ahora, por serie trigonométrica:

$$x(t) = a_0 + \sum_{n=1}^N{a_n\cos(n\omega_0 t) + b_n\sin(n\omega_0 t)},$$

donde:

$$a_0 = c_0 = \frac{1}{tf-ti}\int_{t_i}^{t_f}x(t)dt$$

y

$$a_n = \frac{2}{tf-ti}\int_{t_i}^{t_f}x(t)\cos(n\omega_0 t)dt$$

$$b_n = \frac{2}{tf-ti}\int_{t_i}^{t_f}x(t)\sin(n\omega_0 t)dt$$







Dado que $x(t)$ presenta simetría par, $x(t)=x(-t)$:


 $$\boxed{b_n = 0 \quad \forall n \in \{0,1,\dots,N\}}$$

Así:

$$x(t)=|A\cos(2\pi Fo t)|^2  = \frac{A^2}{2}+\frac{A^2*\cos(4\pi Fo t)}{2} = a_0 + \sum_{n=-N}^N{a_n\cos(n\omega_0 t)}$$

$a_0$ codifica el nivel DC de la señal, entonces:

$$a_0 = \frac{1}{\frac{1}{2F_0}-(-\frac{1}{2F_0}})\int_{-\frac{1}{2F_0}}^{\frac{1}{2F_0}}\left(\frac{A^2}{2}+\frac{A^2*\cos(4\pi Fo t)}{2}\right)dt =\int_{-\frac{1}{2F_0}}^{\frac{1}{2F_0}}(\frac{A^2}{2})dt+\int_{-\frac{1}{2F_0}}^{\frac{1}{2F_0}}\frac{A^2*\cos(4\pi Fo t)}{2}dt$$



$$\boxed{a_0 = \frac{A^2}{2}}$$




## Cálculo de \(a_0\)
El coeficiente \(a_0\) se calcula como:
$$
a_0 = \frac{1}{T} \int_{-T/2}^{T/2} x(t) \, dt
$$
Donde \(T = \frac{1}{F_0}\). Sustituyendo \(x(t)\):
$$
a_0 = \frac{1}{T} \int_{-T/2}^{T/2} \left( \frac{A^2}{2} + \frac{A^2 \cos(4\pi F_0 t)}{2} \right) dt
$$
Separando la integral:
$$
a_0 = \frac{1}{T} \left( \int_{-T/2}^{T/2} \frac{A^2}{2} dt + \int_{-T/2}^{T/2} \frac{A^2 \cos(4\pi F_0 t)}{2} dt \right)
$$
La primera integral es:
$$
\int_{-T/2}^{T/2} \frac{A^2}{2} dt = \frac{A^2}{2} \cdot T
$$
La segunda integral es cero porque la integral de \(\cos(4\pi F_0 t)\) sobre un período completo es cero. Por lo tanto:
$$
a_0 = \frac{1}{T} \cdot \frac{A^2}{2} \cdot T = \frac{A^2}{2}
$$

## Cálculo de \(a_n\)
El coeficiente \(a_n\) se calcula como:
$$
a_n = \frac{2}{T} \int_{-T/2}^{T/2} x(t) \cos(n \omega_0 t) \, dt
$$
Sustituyendo \(x(t)\):
$$
a_n = \frac{2}{T} \int_{-T/2}^{T/2} \left( \frac{A^2}{2} + \frac{A^2 \cos(4\pi F_0 t)}{2} \right) \cos(n \omega_0 t) \, dt
$$
Separando la integral:
$$
a_n = \frac{2}{T} \left( \int_{-T/2}^{T/2} \frac{A^2}{2} \cos(n \omega_0 t) \, dt + \int_{-T/2}^{T/2} \frac{A^2 \cos(4\pi F_0 t)}{2} \cos(n \omega_0 t) \, dt \right)
$$
La primera integral es cero porque la integral de \(\cos(n \omega_0 t)\) sobre un período completo es cero. La segunda integral se resuelve usando la identidad del producto de cosenos:
$$
\cos(A) \cos(B) = \frac{1}{2} [\cos(A+B) + \cos(A-B)]
$$
Aplicando esto:
$$
a_n = \frac{2}{T} \int_{-T/2}^{T/2} \frac{A^2}{2} \cdot \frac{1}{2} [\cos((4\pi F_0 + n \omega_0) t) + \cos((4\pi F_0 - n \omega_0) t)] \, dt
$$
Dado que \(\omega_0 = 2\pi F_0\), esto se simplifica a:
$$
a_n = \frac{A^2}{T} \int_{-T/2}^{T/2} [\cos((4\pi F_0 + n 2\pi F_0) t) + \cos((4\pi F_0 - n 2\pi F_0) t)] \, dt
$$
Las integrales de los términos coseno serán cero a menos que el argumento del coseno sea cero. Esto ocurre cuando \(n = 2\):
$$
a_2 = \frac{A^2}{2}
$$
Para otros valores de \(n\), \(a_n = 0\).

## Cálculo de \(b_n\)
Dado que \(x(t)\) es una función par (\(x(t) = x(-t)\)), los coeficientes \(b_n\) son cero para todo \(n\):
$$
b_n = 0 \quad \forall n
$$

## Resumen
$$
a_0 = \frac{A^2}{2}
$$
$$
a_2 = \frac{A^2}{2}
$$
$$
a_n = 0 \quad \text{para } n \neq 0, 2
$$
$$
b_n = 0 \quad \forall n
$$

```python
import numpy as np  # Importa la librería numpy para operaciones numéricas.
import matplotlib.pyplot as plt  # Importa la librería matplotlib para graficar.

# Definición de parámetros de la señal
A = 8  # Amplitud de la señal.
Fo = 1  # Frecuencia fundamental en Hz.
Fs = 100 * Fo  # Frecuencia de muestreo, 100 veces la frecuencia fundamental.
T = 1 / Fo  # Período de la señal.
ti = -T / 2  # Tiempo inicial para el intervalo de tiempo.
tf = T / 2   # Tiempo final para el intervalo de tiempo.
N = 50  # Número de armónicos a considerar.

wo = 2 * np.pi / T  # Frecuencia angular fundamental.
tv = np.arange(ti, tf, 1 / Fs)  # Vector de tiempo desde ti hasta tf con paso 1/Fs.

# Definir bases de Fourier
phin = np.zeros((len(tv), 2 * N + 1), dtype=np.complex_)  # Inicializa una matriz para almacenar las bases de Fourier.
for n in range(-N, N + 1, 1):
    phin[:, n + N] = np.exp(1j * n * wo * tv)  # Calcula la base exponencial de Fourier para cada armónico n.

# Calcular coeficientes de la serie de Fourier
cn = np.zeros(2 * N + 1, dtype=np.complex_)  # Inicializa un arreglo para los coeficientes de Fourier.
nv = np.linspace(-N, N, 2 * N + 1)  # Vector de armónicos desde -N hasta N.
cn[N] = A**2 / 2  # Coeficiente DC (C_0).
cn[N - 2] = A**2 / 4  # Coeficiente para el armónico -2.
cn[N + 2] = A**2 / 4  # Coeficiente para el armónico +2.

# Definir la señal original
x = abs(A * np.cos(2 * np.pi * Fo * tv)) ** 2  # Señal original x(t).

# Potencia de la señal original
Px = 3 * (A**4) / 8  # Potencia de la señal original.

# Función para graficar la señal original
def plot_original_signal():
    plt.figure(figsize=(8, 6))  # Crea una figura de tamaño 8x6.
    plt.plot(tv, x, color='r', label='$x(t)$ (Señal original)')  # Grafica la señal original.
    plt.title('Señal original $x(t)$', fontsize=16)  # Título del gráfico.
    plt.xlabel('t [s]')  # Etiqueta del eje x.
    plt.ylabel('x(t)')  # Etiqueta del eje y.
    plt.grid(True)  # Muestra la cuadrícula.
    plt.legend()  # Muestra la leyenda.
    plt.show()  # Muestra el gráfico.

# Función para graficar el espectro de Fourier
def plot_spectrum():
    fig, axs = plt.subplots(2, 2, figsize=(10, 8))  # Crea una figura con 4 subgráficos.
    axs[0, 0].stem(nv, np.real(cn), 'r')  # Grafica la parte real de los coeficientes.
    axs[0, 0].set_xlabel(r'$n \cdot w_o$ [rad/s]')  # Etiqueta del eje x.
    axs[0, 0].set_ylabel(r'Re$\{C_n\}$')  # Etiqueta del eje y.
    axs[0, 0].grid(True)  # Muestra la cuadrícula.
    axs[0, 1].stem(nv, np.imag(cn), 'r')  # Grafica la parte imaginaria de los coeficientes.
    axs[0, 1].set_xlabel(r'$n \cdot w_o$ [rad/s]')  # Etiqueta del eje x.
    axs[0, 1].set_ylabel(r'Im$\{C_n\}$')  # Etiqueta del eje y.
    axs[0, 1].grid(True)  # Muestra la cuadrícula.
    axs[1, 0].stem(nv, 20 * np.log10(np.abs(cn)), 'r')  # Grafica la magnitud en dB de los coeficientes.
    axs[1, 0].set_xlabel(r'$n \cdot w_o$ [rad/s]')  # Etiqueta del eje x.
    axs[1, 0].set_ylabel(r'$|C_n|$ [dB]')  # Etiqueta del eje y.
    axs[1, 0].grid(True)  # Muestra la cuadrícula.
    axs[1, 1].stem(nv, np.angle(cn), 'r')  # Grafica la fase de los coeficientes.
    axs[1, 1].set_xlabel(r'$n \cdot w_o$ [rad/s]')  # Etiqueta del eje x.
    axs[1, 1].set_ylabel(r'$\angle C_n$ [rad]')  # Etiqueta del eje y.
    axs[1, 1].grid(True)  # Muestra la cuadrícula.
    fig.tight_layout()  # Ajusta el layout para evitar superposiciones.
    plt.show()  # Muestra el gráfico.

# Función para graficar la reconstrucción de la señal
def plot_reconstruccion(Na):
    ind = range(N - Na, N + Na + 1)  # Índices de los armónicos a considerar.
    er = 1 - np.sum(np.abs(cn[ind]) ** 2) / Px  # Calcula el error de reconstrucción.
    xe = phin[:, ind].dot(cn[ind])  # Reconstruye la señal usando los coeficientes seleccionados.
    plt.figure(figsize=(8, 6))  # Crea una figura de tamaño 8x6.
    plt.plot(tv, x, color='r', label='$x(t)$ (Original)')  # Grafica la señal original.
    plt.plot(tv, xe, color='b', label='$x_e(t)$ (Reconstruida)')  # Grafica la señal reconstruida.
    plt.title(f'E_r = {100 * er:.2f}% (Error de reconstrucción)', fontsize=16)  # Título del gráfico con el error.
    plt.xlabel('t [s]')  # Etiqueta del eje x.
    plt.ylabel('x(t)')  # Etiqueta del eje y.
    plt.grid(True)  # Muestra la cuadrícula.
    plt.legend()  # Muestra la leyenda.
    plt.show()  # Muestra el gráfico.

# Mostrar resultados
graph_Na = 5  # Número de armónicos para la reconstrucción.
plot_original_signal()  # Llama a la función para graficar la señal original.
plot_spectrum()  # Llama a la función para graficar el espectro de Fourier.
plot_reconstruccion(graph_Na)  # Llama a la función para graficar la reconstrucción de la señal.
```

# **Distorsión Total de Armónicos (THD) y Factor de Potencia en un Circuito Eléctrico**

## 1. ¿En qué consiste la distorsión total de armónicos (THD)?

La **Distorsión Total de Armónicos (THD)** es una métrica que mide el contenido de armónicos en una señal con respecto a su componente fundamental. Se define como:

$$
THD = \frac{\sqrt{\sum_{n=2}^{\infty} V_n^2}}{V_1} \times 100\%
$$

donde:

- \(V_1\) es la amplitud del componente fundamental (primer armónico).
- \(V_n\) son las amplitudes de los armónicos de orden \(n\) (con \(n > 1\)).

Un THD elevado indica que la señal tiene alta distorsión armónica, lo que puede ocasionar pérdidas en sistemas eléctricos y afectar la eficiencia de los dispositivos.

---

## 2. ¿Qué es el factor de potencia y cómo se relaciona con el THD?

El **factor de potencia (FP)** mide la eficiencia con la que un circuito eléctrico convierte la energía en trabajo útil. Se expresa como:

$$
FP = \cos \varphi
$$

donde $$( \varphi )$$ es el ángulo de fase entre la corriente y la tensión.

Cuando hay distorsión armónica, el factor de potencia total se ve afectado y puede calcularse en función del THD:

$$
FP_{dist} = \frac{1}{\sqrt{1 + THD^2}}
$$

Esto muestra que, a medida que aumenta el THD, el factor de potencia disminuye, reduciendo la eficiencia del sistema.

---

## 3. ¿Cómo calcular el THD desde la FFT?

La **Transformada Rápida de Fourier (FFT)** permite descomponer la señal en sus componentes de frecuencia. Para calcular el THD:

1. Se obtiene la FFT de la señal de salida.
2. Se identifican las amplitudes de los armónicos.
3. Se aplica la ecuación:

$$
THD = \frac{\sqrt{\sum_{n=2}^{\infty} |X(f_n)|^2}}{|X(f_1)|}
$$

donde:

- \( |X(f_1)| \) es la magnitud de la componente fundamental.
- \( |X(f_n)| \) son las magnitudes de los armónicos superiores.

En el código ilustrativo, esta operación se realiza en la función `calculate_thd(signal)`.

---

## 4. Cálculo de la distorsión del factor de potencia a partir del THD

El factor de potencia considerando la distorsión armónica se obtiene con:

$$
FP_{dist} = \frac{1}{\sqrt{1 + THD^2}}
$$

En el código ilustrativo, esta operación se realiza en la función `calculate_power_factor(THD)`.








```python
import numpy as np
import scipy.signal as sig
import matplotlib.pyplot as plt
import sympy as sym


# Parámetros de la señal de entrada
A = 120  # Amplitud de la señal senoidal
Fo = 60   # Frecuencia de la señal en Hz
Fs = 30 * Fo  # Frecuencia de muestreo
To = 1 / Fo   # Periodo fundamental
Ts = 1 / Fs   # Periodo de muestreo
t = np.arange(0, 5 * To, Ts)  # Simulación de 5 periodos

# Selección del circuito con validación
while True:
    tipo_circuito = input("Seleccione el tipo de circuito (Resistivo o RC): ").strip().lower()
    if tipo_circuito in ["resistivo", "rc"]:
        break
    print("Entrada no válida. Escriba 'Resistivo' o 'RC'.")

# Ingreso de valores con conversión de unidades
R_v = float(input("Ingrese el valor de la resistencia R en ohmios: "))

if tipo_circuito == "rc":
    C_v_micro = float(input("Ingrese el valor de la capacitancia C en microfaradios (µF): "))
    C_v = C_v_micro * 1e-6  # Convertir de µF a F
    X_C = 1 / (2 * np.pi * Fo * C_v)  # Reactancia capacitiva
    Z = np.sqrt(R_v**2 + X_C**2)  # Impedancia total
    FP = R_v / Z  # Factor de potencia correcto
    print(f"Factor de Potencia del circuito RC serie: {FP:.2f}")
else:
    C_v = None
    FP = 1.0  # Para un circuito puramente resistivo, FP = 1
    print(f"Factor de Potencia del circuito Resistivo: {FP:.2f}")

# Generación de la señal de entrada (rectificada completa)
in_o = A * np.sin(2 * np.pi * Fo * t)  # Señal senoidal
in_ = np.abs(in_o)  # Señal rectificada

# Cálculo de la respuesta del circuito
if C_v is None:
    out = in_  # La salida es la misma entrada para el circuito resistivo
else:
    # Transferencia del circuito RC
    num = np.array([1])
    den = np.array([R_v * C_v, 1])
    G_n = sig.TransferFunction(num, den)
    _, out, _ = sig.lsim(G_n, in_, t)

# Gráfica de la señal de entrada y salida
plt.figure(figsize=(10, 5))
plt.plot(t, in_, label='Entrada (Rectificada)')
plt.plot(t, out, label='Salida del Circuito')
plt.xlabel('Tiempo [s]')
plt.ylabel('Voltaje [V]')
plt.legend()
plt.title(f'Señales de Entrada y Salida - Circuito {tipo_circuito.capitalize()}')
plt.grid()
plt.show()

# Cálculo de la FFT para la señal de salida
vfre = np.fft.rfftfreq(len(out), 1 / Fs)
Xf = np.fft.rfft(out)

plt.figure(figsize=(10, 5))
markerline, stemlines, baseline = plt.stem(vfre, np.abs(Xf) / len(out))
plt.setp(stemlines, linewidth=1)  # Ajustar grosor de las líneas
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('|X(f)|')
plt.title('Espectro de Frecuencia (FFT) de la Salida')
plt.grid()
plt.show()

# Cálculo del THD
def calculate_thd(signal):
    N = len(signal)
    Y = np.fft.rfft(signal)
    Y = 2.0 / N * np.abs(Y[:N // 2])
    fundamental = Y[1]  # Primer armónico
    harmonics = np.sqrt(np.sum(Y[2:]**2))  # Suma de componentes armónicas
    THD = harmonics / fundamental
    return THD

THD = calculate_thd(out)
print(f"THD (Distorsión Armónica Total) para el circuito {tipo_circuito.capitalize()}: {THD:.2%}")
```

##  Discusión de resultados

- **Para la carga resistiva**, la salida es una señal rectificada sin filtrar, lo que implica una mayor cantidad de armónicos y un **THD alto**.
- **Para la carga RC**, el capacitor filtra la salida, reduciendo la presencia de armónicos y disminuyendo el **THD**, lo que mejora el **factor de potencia**.
- **Patrón esperado:**
  - **Sin filtro:** THD elevado, FP bajo.
  - **Con filtro (RC):** THD disminuye, FP aumenta.
- **Efecto de variar \( R \) y \( C \):**
  - Aumentar \( C \) mejora el filtrado, disminuyendo el THD y mejorando el factor de potencia.
  - Aumentar \( R \) disminuye la corriente, pero puede no tener un efecto notable en la salida.

Esto es clave en aplicaciones donde la calidad de la energía sea importante, como en fuentes de alimentación y sistemas de distribución eléctrica.
