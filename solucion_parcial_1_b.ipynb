```python
import numpy as np
import matplotlib.pyplot as plt
#Definir parámetros de la simulación

#Periodos de la señal
T1 = 1/500
T2 = 1/1000
T3 = 1/5500
#Periodo señal cuasiperiodica
T = 1/5500
#Frecuencias
F1 = 1/T1
F2 = 1/T2
F3 = 1/T3
#Fs: frecuencia muestreo
Fs = 40 * F3
Ts = 1/Fs #periodo muestreo

t = np.arange(0,2*T,Ts) #vector de tiempo que incluye dos periodos cuasiperiódica
x = 3*np.cos(2*np.pi*F1*t) + 5*np.sin(2*np.pi*F2*t) + 10*np.cos(2*np.pi*F3*t)

plt.plot(t,x)
plt.xlabel('$t[s]$')
plt.ylabel('$x(t)$')
plt.title('$x(t)=3\cos(1000*pi*t) + 5 \sin(2000*pi*t) + 10\cos(11000*pi*t)$')
plt.show()
```
